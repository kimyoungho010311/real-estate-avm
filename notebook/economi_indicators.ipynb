```python
import requests
import json
import os
from datetime import datetime
import time

# 기본 파라미터
KEY = '4afaa363f1b3428c88d1660d5e70208c'
STATBL_ID = 'A_2024_00554'
DTACYCLE_CD = 'MM'
CLS_ID = '510025'

# 시작, 종료 월
start_date = datetime(2000, 1, 1)
end_date = datetime(2025, 9, 1)

current = start_date
while current <= end_date:
    yyyymm_api = current.strftime('%Y%m')   # API 요청용 (하이픈 없음)
    yyyymm_dir = current.strftime('%Y-%m')  # 폴더명 저장용 (하이픈 있음)

    url = (
        f'https://www.reb.or.kr/r-one/openapi/SttsApiTblData.do'
        f'?KEY={KEY}&Type=json&STATBL_ID={STATBL_ID}'
        f'&DTACYCLE_CD={DTACYCLE_CD}&CLS_ID={CLS_ID}'
        f'&START_WRTTIME={yyyymm_api}&END_WRTTIME={yyyymm_api}'
    )

    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        data = response.json()
    except Exception as e:
        print(f'[{yyyymm_api}] 요청 실패: {e}')
        # 다음 달로 이동
        if current.month == 12:
            current = current.replace(year=current.year+1, month=1)
        else:
            current = current.replace(month=current.month+1)
        time.sleep(1)
        continue

    # 저장 디렉토리 (dt=yyyy-mm)
    save_dir = os.path.join('data', f'dt={yyyymm_dir}')
    os.makedirs(save_dir, exist_ok=True)

    # 파일 저장
    save_path = os.path.join(save_dir, f'{yyyymm_dir}.json')
    with open(save_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

    print(f'[{yyyymm_dir}] 저장 완료')

    # 다음 달로 이동
    if current.month == 12:
        current = current.replace(year=current.year+1, month=1)
    else:
        current = current.replace(month=current.month+1)

    # 서버 부하 방지
    time.sleep(0.5)
```
